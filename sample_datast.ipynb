```python
import pandas as pd


artists = pd.read_csv("archive/artists-data.csv")
lyrics = pd.read_csv("archive/lyrics-data.csv")
```

```python
l = lyrics.query('ALink == "/taylor-swift/"')

songs_lyrics = l["Lyric"]
# split these lyrics by new line character for each song
final_split = []
for i in songs_lyrics:
    final_split.append(i.split("\n"))
# final_split is a list of lists make it just a list
final_split = [item for sublist in final_split for item in sublist]
# remove empty strings from the list    
final_split = [x for x in final_split if x != '']
final_split
```

```python
import datasets

ds = datasets.Dataset.from_dict({'text': songs_lyrics})
ds.save_to_disk("./sample_dataset")
# ds.push_to_hub("BhavyaMuni/sample_dataset")
```
